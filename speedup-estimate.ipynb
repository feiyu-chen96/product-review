```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cores = np.array([1, 2, 3, 4])

preprocess = np.array([43.609, 38.570, 36.640, 36.224])
train = np.array([73.580, 55.666, 53.673, 51.367])
predict = np.array([44.087, 33.050, 32.803, 29.785])
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(121)
plt.plot(cores, preprocess, linestyle='--', marker='x', label='preprocess')
plt.plot(cores, train, linestyle='--', marker='x', label='train')
plt.plot(cores, predict, linestyle='--', marker='x', label='predict')
plt.ylim(0, 80)
plt.legend()
plt.xlabel("number of cores")
plt.ylabel("execution time(s)")
plt.title("execution time of 3 phases");

plt.subplot(122)
plt.plot(cores, preprocess[0]/preprocess, linestyle='--', marker='x', label='preprocess')
plt.plot(cores, train[0]/train, linestyle='--', marker='x', label='train')
plt.plot(cores, predict[0]/predict, linestyle='--', marker='x', label='predict')
plt.ylim(0, 2)
plt.legend()
plt.xlabel("number of cores")
plt.ylabel("speedup")
plt.title("speedup of 3 phases");
```
